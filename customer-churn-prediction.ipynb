```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
churn = pd.read_csv('/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv')
churn.head()
```

```python
churn = churn.drop(['customerID'],axis =1)
churn.head()
```

```python
churn = pd.get_dummies(churn,drop_first=True)
churn.head()
```

```python
x=churn.iloc[:,:-1]
x
```

```python
y=churn.iloc[:,-1]
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state = 20)
```

```python
y_train
```

```python
rf_clf = RandomForestClassifier(random_state=20)
```

```python
rf_clf = rf_clf.fit(x_train, y_train)
```

```python
test_pred = rf_clf.predict(x_test)
test_pred
```

```python
print(classification_report(test_pred, y_test))
```

```python
dt_clf = DecisionTreeClassifier(random_state=20)
```

```python
dt_clf = dt_clf.fit(x_train, y_train)
```

```python
test_pred1 = dt_clf.predict(x_test)
test_pred1
```

```python
print(classification_report(test_pred1,y_test))
```
